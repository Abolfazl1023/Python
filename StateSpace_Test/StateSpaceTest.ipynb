### Problem #1

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
# D.E.: 21y(2)+14y(1)+7y=7u
A=np.array([[0.0,1.0],[-1.0,-2.0]])
B=np.array([[0.0,1.0]]).T
C=np.array([[1.0,0.0]])
D=np.array([[0.0]])

sys1=signal.StateSpace(A,B,C,D)
t1, y1 = signal.step(sys1) # Step Response
plt.plot(t1,y1)




```

#### Problem #2

```python
# D.E.: 21y(2)+14y(1)+7y=7u

A=np.array([[0.0,1.0],[-1.0,-2.0]])
B=np.array([[0.0,1.0]]).T
C=np.array([[1.0,0.0]])
D=np.array([[0.0]])

# Eigen Values
E=np.linalg.eig(A)[0]
sys2=signal.StateSpace(A,B,C,D)
t=np.linspace(0,50,51)
u=np.zeros(len(t))
u[0:25]=1.0
u[25:]=2.0
t2, y2, x2 = signal.lsim(sys2,u,t)



print('Eigen Values:', E)
plt.plot(t,u, label="input")
plt.plot(t2,y2, label="output")
plt.legend(loc='lower right')
```
