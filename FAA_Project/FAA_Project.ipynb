## Since the number of columns cause a group function by ACFT_NAME
## are manifold therefore, the result are not good enough. In fact, we can't see 
## any group

```python
grp_by_ACName = selected_data.groupby(['LOC_STATE_NAME','ACFT_MODEL_NAME','RMK_TEXT','FLT_PHASE','EVENT_TYPE_DESC','FATAL_FLAG'])['ACFT_MAKE_NAME'].apply(', '.join).reset_index()
pd.set_option('display.max_row',100)
grp_by_ACName
```
